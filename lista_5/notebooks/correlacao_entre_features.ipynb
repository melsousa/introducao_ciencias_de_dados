```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
telegram = pd.read_csv('../dataset/mensagens_com_frequencia_e_texto_processado.csv')
telegram.head()
```

```python
telegram['frequencia'].describe()
```

```python
from collections import Counter

# Contar palavras (split por espaço)
telegram['qtd_palavras'] = telegram['texto_original'].str.split().str.len()

# Contar caracteres (inclui espaços e pontuação)
telegram['qtd_caracteres'] = telegram['texto_original'].str.len()

# Contar a frequência de palavras
word_freq = Counter(" ".join(telegram['texto_original']).split())

# Adicionar a frequência de palavras como uma nova coluna
telegram['word_freq'] = telegram['texto_original'].apply(lambda x: sum(word_freq[word] for word in x.split()))
```

```python
# Contar palavras (split por espaço)
telegram['qtd_palavras'] = telegram['texto_original'].str.split().str.len()

# Contar caracteres (inclui espaços e pontuação)
telegram['qtd_caracteres'] = telegram['texto_original'].str.len()

telegram['media_palavra'] = telegram['qtd_caracteres'] / telegram['qtd_palavras']
telegram['frequencia_relativa'] = telegram['frequencia'] / telegram['qtd_palavras']
telegram['intensidade_texto'] = telegram['qtd_palavras'] * telegram['frequencia']

telegram.head()
```

```python


correlation = telegram[['qtd_palavras', 'qtd_caracteres', 'word_freq', 'media_palavra', 'frequencia_relativa','intensidade_texto', 'frequencia']].corr()

correlation
```
